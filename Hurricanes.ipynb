*** Janitorial Services using regex ***

```python
import re
from decimal import Decimal

def gettabledata(data_value):
    data_value = re.sub("\n", u'', data_value)
    data_value = re.sub(",", u'', data_value)
    data_value = re.sub(":", u'', data_value)
    data_value = re.sub("[+]", u'', data_value)
    data_value = re.sub("[$]", u'', data_value)
    data_value = re.sub("~", u'', data_value)  
    data_value = re.sub("<", u'', data_value)
    data_value = re.sub(">", u'', data_value)
    data_value = re.sub(" ", u'', data_value)
    data_value = re.sub("\xa0", u'', data_value)
    data_value = re.sub("Unknown", u'', data_value)
    data_value = re.sub("Numerous", u'', data_value)
    data_value = re.sub("Not", u'', data_value)
    data_value = re.sub("known", u'', data_value)
    data_value = re.sub("None", u'', data_value)
    data_value = re.sub(u'\u2265', u'', data_value)
    return data_value

d = {
      'million': 6,
      'billion': 9
    }
    

def million_billion(data_value):
    if data_value[-7:]in d:
        num, magnitude = data_value[:-7], data_value[-7:]
        return str(Decimal(num) * 10 ** d[magnitude])
    else:
        return data_value

```

# parse the table and append data to dataframe

```python
from bs4 import BeautifulSoup
import urllib2
import pandas as pd
wiki = "https://en.wikipedia.org/wiki/Atlantic_hurricane_season"
header = {'User-Agent': 'Mozilla/5.0'} #Needed to prevent 403 error on Wikipedia
req = urllib2.Request(wiki,headers=header)
page = urllib2.urlopen(req)
soup = BeautifulSoup(page)
 
year = ""
tropical_storms = ""
hurricanes = ""
major_hurricanes = ""
deaths = ""
damage = ""
notes = ""

i=0

tables = soup.findAll("table", { "class" : "wikitable sortable" })

df = pd.DataFrame(columns=('year', 'tropical_storm', 'hurricanes','major_hurricanes', 'deaths', 'damage','notes' ))
for table in tables:
    rows = table.findAll('tr')
    for row in rows:
        cells = row.find_all('td')
    #For each "tr", assign each "td" to a variable.
        if len(cells) == 6:
            year =  gettabledata(cells[0].find(text=True))
            tropical_storms = gettabledata(cells[1].find(text=True))
            hurricanes = gettabledata(cells[2].find(text=True))
            major_hurricanes = gettabledata(cells[3].find(text=True))
            deaths = gettabledata(cells[4].find(text=True))
            notes = gettabledata(cells[5].find(text=True))

            df.loc[i] = [year,tropical_storms , hurricanes,major_hurricanes , deaths, '', notes ]
            i+=1

        if len(cells) == 7:
            year = gettabledata(cells[0].find(text=True))
            tropical_storms = gettabledata(cells[1].find(text=True))
            hurricanes = gettabledata(cells[2].find(text=True))
            major_hurricanes = gettabledata(cells[3].find(text=True))
            deaths = gettabledata(cells[4].find(text=True))
            notes = gettabledata(cells[5].find(text=True))

            df.loc[i] = [year,tropical_storms , hurricanes,major_hurricanes , deaths, '', notes ]
            i+=1


        if len(cells) == 8:
            year = gettabledata(cells[0].find(text=True))
            tropical_storms = gettabledata(cells[1].find(text=True))
            hurricanes = gettabledata(cells[2].find(text=True))
            major_hurricanes = gettabledata(cells[3].find(text=True))
            deaths = gettabledata(cells[4].find(text=True))
            damage = gettabledata(cells[5].find(text=True))
            damage = million_billion(damage)
            notes = gettabledata(cells[6].find(text= True))

            df.loc[i] = [year,tropical_storms , hurricanes,major_hurricanes , deaths, damage, notes ]
            i+=1


        if len(cells) == 9:
            year = gettabledata(cells[0].find(text=True))
            tropical_storms = gettabledata(cells[1].find(text=True))
            hurricanes = gettabledata(cells[2].find(text=True))
            major_hurricanes = gettabledata(cells[3].find(text=True))
            deaths = gettabledata(cells[4].find(text=True))
            damage = gettabledata(cells[5].find(text=True))
            damage = million_billion(damage)
            notes = gettabledata(cells[6].find(text= True))

            df.loc[i] = [year,tropical_storms , hurricanes,major_hurricanes , deaths, damage, notes ]
            i+=1

        if len(cells) == 10:
            year = gettabledata(cells[0].find(text=True))
            tropical_storms = gettabledata(cells[2].find(text=True))
            hurricanes = gettabledata(cells[3].find(text=True))
            major_hurricanes = gettabledata(cells[4].find(text=True))
            deaths = gettabledata(cells[5].find(text=True))
            damage = gettabledata(cells[6].find(text=True))
            damage = million_billion(damage)
            notes = gettabledata(cells[7].find(text= True))

            df.loc[i] = [year,tropical_storms , hurricanes,major_hurricanes , deaths, damage, notes ]
            i+=1           


    
            
      
 

            
    
            




```

*** Write to the database***

```python
import sqlite3
conn = sqlite3.connect("hurricanes.db")
df.to_sql("hurricanes", conn, index=False)
```

# 1. What year had the most tropical storms?

```python
print(conn.execute("SELECT year, MAX(tropical_storm) FROM hurricanes;").fetchall())
```

# 2. What year had the most major hurricanes?

```python
print(conn.execute("SELECT year, MAX(major_hurricanes) FROM hurricanes;").fetchall())
```

# 3. What year had the most deaths?

```python
print(conn.execute("SELECT year, MAX(deaths) FROM hurricanes;").fetchall())
```

# 4. What year had the most damage (not inflation adjusted)?

```python
print(conn.execute("SELECT year, MAX(damage) FROM hurricanes;").fetchall())
```

*** What year had the highest proportion of tropical storms turn into major hurricanes? ***

```python
print(conn.execute("SELECT year, MAX(CAST(major_hurricanes AS FLOAT)/CAST(tropical_storm AS FLOAT)) FROM hurricanes;").fetchall())
```
